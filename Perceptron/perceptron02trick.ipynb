## **Perceptron Trick**

```python
from sklearn.datasets import make_classification
import numpy as np
```

## **Generating Dataset**

```python
x,y = make_classification(n_samples= 100, n_features=2, n_informative=1, n_redundant=0, n_classes=2,
                          n_clusters_per_class=1, class_sep=10, random_state=41, hypercube=False)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x[:,0],x[:,1],c=y,cmap='winter',s=100)
```

```python
#we add 1 with co-ordinate
np.insert(x,0,1,axis=1)
```

```python
def step(z):
    return 1 if z>0 else 0
```

```python
def perceptron(X,y):

    # adding 1 with co-ordinate 
    X = np.insert(X,0,1,axis=1)

    #weights according to shape of x co-ordinate --> eg: input col is 3 and weight is 3
    weights = np.ones(X.shape[1])

    #learning rate
    lr = 0.1

    for i in range(1000):
        #find random value of x
        j = np.random.randint(0,100)

        #predict output
        y_hat = step(np.dot(X[j],weights))

        #calculate new weight
        weights = weights + lr*(y[j] - y_hat)*X[j]

        #return intercept and coef_
    return weights[0],weights[1:]
```

```python
intercept_, coef_ = perceptron(x,y)
```

```python
print(coef_)
print(intercept_)
```

```python
m =  -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m * x_input + b
```

```python
#plot graph
plt.figure(figsize=(10,5))
plt.plot(x_input,y_input,color='red',linewidth = 3)
plt.scatter(x[:,0],x[:,1],c=y,cmap='winter',s=100)
plt.ylim(-4,3)
```
