```python
#remove columns
df.drop(columns = ['RowNumber','CustomerId', 'Surname'],inplace = True)
```

```python
df.head()
```

```python
df = pd.get_dummies(df, columns = ['Geography', 'Gender'],dtype = int, drop_first = True)
```

```python
df.head()
```

```python
#split data 
from sklearn.model_selection import train_test_split
X = df.drop(columns = ['Exited'])
y = df['Exited']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2, random_state = 1)
```

```python
X_train
```

```python
y_train.shape
```

```python
#scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled.shape
```

```python
#importing keras
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model  = Sequential()

#hidden layer -> 3 input node
model.add(Dense(3,activation = 'sigmoid',input_dim = 11))

#output layer
model.add(Dense(1,activation = 'sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss = 'binary_crossentropy',optimizer = 'Adam')
```

```python
model.fit(X_train_scaled, y_train, epochs = 10)
```

```python
#weights
model.layers[0].get_weights()
```

```python
model.layers[1].get_weights()
```

```python
#prediction 
y_log = model.predict(X_test_scaled)
```

```python
#prediction in probability form. We have to convert this into 0 and 1.
y_pred = np.where(y_log > 0.5,1,0)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

### **Imporve Mode**
1. Increase number of epochs
2. Activation function =  Relu
3. Increase number of nodes in hidden layer. current time we have only 3 nodes
4. Increase number of layers

```python
#hidden layer -> 11 input node
model.add(Dense(11,activation = 'relu',input_dim = 11))

#hidden layer -> 11 input node
model.add(Dense(11, activation = 'relu'))

#output layer
model.add(Dense(1,activation = 'sigmoid'))
```

```python
model.compile(loss = 'binary_crossentropy', optimizer='Adam', metrics = ['accuracy'])
```

```python
''' validation_split -> split train data into 80% and 20% where 80% data will train and 
20% data will test and show accuray'''
history = model.fit(X_train_scaled, y_train, epochs = 100, validation_split = 0.2)
```
